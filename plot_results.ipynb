```python
data = pd.read_csv('./results/results.csv')
```

```python
pd_pf_opt = data['pd_pf_opt']
pd_pf = data['pd_pf']
pd_pomcp = data['pd_pomcp']

rmse_pf = data['rmse_pf']
rmse_pf_opt = data['rmse_pf_opt']
rmse_pomcp = data['rmse_pomcp']

rmse_velocity_pf_opt = data['rmse_velocity_pf_opt']
rmse_velocity_pf = data['rmse_velocity_pf']
rmse_velocity_pomcp = data['rmse_velocity_pomcp']
snr_values = data['snr_values']

fig = plt.figure()
plt.plot(pd_pf_opt, label = 'Particle Filtre Opt')
plt.plot(pd_pf, label = 'Particle Filtre')
plt.plot(pd_pomcp, label = 'POMCP')

plt.legend()
plt.xlabel('Time (s)')
plt.ylabel("Probability of detection")
plt.title('Probability of detection')
plt.savefig('./results/plots/proba_detection.png')
plt.close(fig)

fig = plt.figure()
plt.plot(rmse_pf_opt, label = 'Particle Filtre Opt')
plt.plot(rmse_pf, label = 'Particle Filtre')
plt.plot(rmse_pomcp, label = 'POMCP')

plt.legend()
plt.xlabel('Time (s)')
plt.ylabel('RMSE (km)')
plt.title('RMSE between the predicted coardinates of each algorithm and ground truth')
plt.savefig('./results/plots/mse_coardinates.png')
plt.close(fig)




fig = plt.figure()
plt.plot(rmse_velocity_pf_opt, label = 'Particle Filtre Opt')
plt.plot(rmse_velocity_pf, label = 'Particle Filtre')
plt.plot(rmse_velocity_pomcp, label = 'POMCP')

plt.legend()
plt.xlabel('Time (s)')
plt.ylabel('RMSE (km)')
plt.title('RMSE between the predicted velocities of each algorithm and ground truth')
plt.savefig('./results/plots/mse_velocities.png')
plt.close(fig)



fig = plt.figure()
plt.plot(snr_values, rmse_velocity_pf_opt, label = 'Particle Filtre Opt')
plt.plot(snr_values, rmse_velocity_pf, label = 'Particle Filtre')
plt.plot(snr_values, rmse_velocity_pomcp, label = 'POMCP')

plt.legend()
plt.xlabel('SNR (dB)')
plt.ylabel('RMSE (km)')
plt.title('RMSE of velocities vs SNR of the target')
plt.savefig('./results/plots/mse_velocities_snr.png')
plt.close(fig)


fig = plt.figure()
plt.plot(snr_values, rmse_pf_opt, label = 'Particle Filtre Opt')
plt.plot(snr_values, rmse_pf, label = 'Particle Filtre')
plt.plot(snr_values, rmse_pomcp, label = 'POMCP')

plt.legend()
plt.xlabel('SNR (dB)')
plt.ylabel('RMSE (km)')
plt.title('RMSE of coordinates vs SNR of the target')
plt.savefig('./results/plots/mse_coordinates_snr.png')
plt.close(fig)


fig = plt.figure()
plt.plot(snr_values, pd_pf_opt, label = 'Particle Filtre Opt')
plt.plot(snr_values, pd_pf, label = 'Particle Filtre')
plt.plot(snr_values, pd_pomcp, label = 'POMCP')

plt.legend()
plt.xlabel('SNR (dB)')
plt.ylabel('RMSE (km)')
plt.title('Probability of detection vs SNR of the target')
plt.savefig('./results/plots/proba_detection_snr.png')
plt.close(fig)

```
